```python
import numpy as np
import cv2
import os
import shutil
from pathlib import Path
import cv2
from skimage import io
from IPython.display import Image, display
import dlib
from matplotlib import pyplot as plt
```

```python
filename = 'Christopher_Conyers_0001.jpg'
```

```python
image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
crop_length = 60
#target_size = image.shape[0] - crop_length
x0 = int((crop_length)/2)
y0 = int((crop_length)/2)
x1 = -x0
y1 = -y0
cr_image = image[x0:x1, y0:y1]

plt.imshow(cr_image)
plt.show()
```

```python
cifar_10_size = 32
scale_size = cifar_10_size
sm_image = cv2.resize(cr_image, (scale_size,scale_size))

plt.imshow(sm_image)
plt.show()
```
