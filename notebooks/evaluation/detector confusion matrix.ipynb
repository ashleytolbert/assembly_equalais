```python
import sys
sys.path.append('../../source')

from pathlib import Path
import cv2
from skimage import io
from IPython.display import Image, display
import dlib
import pandas as pd

from data_processing import import_images_from_directories
from face_detectors import dlib_face_detector, cv2_cascade_face_detector
```

```python
# dataset_path = '../../data/cifar-11'
dataset_path = '../../data/private_proof_v1'
```

```python
detectors = {
    'dlib': dlib_face_detector.DlibFaceDetector(), 
    'cv2': cv2_cascade_face_detector.CV2CascadeFaceDetector()
}
```

```python
def detections(detectors, class_name, max_images_per_class=5000):
    X, _ = import_images_from_directories(dataset_path, class_labels=[class_name], resize_shape=(250, 250), max_images_per_class=max_images_per_class)
    num_dets = {}
    for name, d in detectors.items():
        num_dets[name] = d.contains_a_face(X).sum()
    return num_dets, X.shape[0]
```

```python
index = pd.Index(['dlib_face_detections', 'cv2_face_detections', 'total_class_images'])
confusion = pd.DataFrame(index=index, columns=[p.name for p in list(Path(dataset_path).iterdir()) if 'holdout' not in p.name])
for c in confusion.columns:
    num_dets, total = detections(detectors, c)
    for key, val in num_dets.items():
        confusion.loc[key + '_face_detections', c] = 100.0 * val / total
        confusion.loc['total_class_images', c] = total
```
