```python
import requests
import io
import base64
from PIL import Image as pil_image
import matplotlib.pyplot as plt
```

```python
endpoint_url = 'http://127.0.0.1:8080/'
# endpoint_url = 'https://equal-ais.appspot.com'
```

```python
r = requests.get(endpoint_url)
print(r.content.decode("utf-8"))
```

```python
r = requests.put(endpoint_url)
print(r.content.decode("utf-8"))
```

```python
r = requests.post(endpoint_url)
print(r.content.decode("utf-8"))
```

```python
files = {
    'ignored_key_1': "I'm a string, not an image!", 
    'ignored_key_2': "3"
}
r = requests.post(endpoint_url, files=files)
print(r.content.decode("utf-8"))
```

```python
# files = {'image': open('/home/josh/Downloads/download (1).jpeg', 'rb')}
files = {'image': open('/home/josh/Downloads/image (2).png', 'rb')}
r = requests.post(endpoint_url, files=files)
pil_image.open(io.BytesIO(base64.b64decode(r.content)))
```
